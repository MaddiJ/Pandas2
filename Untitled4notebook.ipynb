```python
# Dependencies and Setup
import pandas as pd
import pathlib as path
```

```python
# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv("schools_complete.csv")
student_data = pd.read_csv("students_complete.csv")

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Total the school and student count -seperately
school_count= len(school_data_complete["school_name"].unique())
student_count= school_data_complete["Student ID"].count()
```

```python
#Total Budget
total_budget = school_data["budget"].sum()
```

```python
#Calculate math, reading, and the average of the two
avg_math_score = school_data_complete["math_score"]>70
avg_reading_score = school_data_complete["reading_score"]>70
avg_passing = (avg_math_score & avg_reading_score).mean()
```

```python
#Calculate passing percent rates
passing_math_count = school_data_complete[(school_data_complete["math_score"] > 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] > 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
```

```python
#Now, let's clean this up
district_summary = pd.DataFrame({"Total Schools": [school_count], 
                                 "Total Students": [student_count], 
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [avg_math_score], 
                                 "Average Reading Score": [avg_reading_score],
                                 "% Passing Math": [passing_math_percentage],
                                 "% Passing Reading": [passing_reading_percentage],
                                 "% Overall Passing Rate": [avg_passing]})

district_summary = district_summary[["Total Schools", "Total Students", "Total Budget",
                                     "Average Math Score", 
                                     "Average Reading Score",
                                     "% Passing Math",
                                     "% Passing Reading",
                                     "% Overall Passing Rate"]]
```

```python
# Students and  Budget district summaries
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
```

```python
#Start of School summary section
```

```python
#Determine the school type
school_types = school_data.set_index(["school_name"])["type"]
```

```python
# Total student count
per_sch_count = school_data_complete["school_name"].value_counts()
```

```python
# Calculate total school budget and per capita spending
per_school_budget = school_data_complete.groupby(["school_name"]).mean(numeric_only=True)["budget"]
per_school_cap = per_school_budget / per_school_count
```

```python
#Average test scores
per_school_math = school_data_complete.groupby(["school_name"]).mean(numeric_only=True)["math_score"]
per_school_reading = school_data_complete.groupby(["school_name"]).mean(numeric_only=True)["reading_score"]
```

```python
#Calculate passing scores by creating a filtered dataframe 

sch_passing_math = school_data_complete[(school_data_complete["math_score"] > 70)]
sch_passing_reading = school_data_complete[(school_data_complete["reading_score"] > 70)]
per_sch_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_count * 100
per_sch_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_count * 100
overall_passing_rate = (per_sch_passing_math + per_sch_passing_reading) / 2
```

```python
#Shall we convert to a dataframe? We shall
per_sch_summary = pd.DataFrame({"School Type": school_types,
                                   "Total Students": per_sch_count,
                                   "Total School Budget": per_school_budget,
                                   "Per Student Budget": per_school_capita,
                                   "Average Math Score": per_school_math,
                                   "Average Reading Score": per_school_reading,
                                   "% Passing Math": per_sch_passing_math,
                                   "% Passing Reading": per_sch_passing_reading,
                                   "% Overall Passing Rate": overall_passing_rate})
```

```python
#Wrangle that data
per_sch_summary = per_school_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",
                                         "Average Math Score", "Average Reading Score", 
                                         "% Passing Math", "% Passing Reading", 
                                         "% Overall Passing Rate"]]

```

```python
# Create Per student budget
per_sch_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].astype(float)
per_sch_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
```

```python
# SCORES BY school spending
# Establish the bins 
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Categorize the spending based on the bins
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_school_cap, spending_bins, labels=group_names)

spending_math_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["Average Math Score"]
spending_reading_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["Average Reading Score"]
spending_passing_math = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Passing Math"]
spending_passing_reading = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Passing Reading"]
overall_passing_rate = (spending_math_scores + spending_reading_scores) / 2
```

```python
# Assemble into data frame
spending_summary = pd.DataFrame({"Average Math Score" : spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Math": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate})
```

```python
 #Wrangle/clean
spending_summary = spending_summary[["Average Math Score", 
                                     "Average Reading Score", 
                                     "% Passing Math", "% Passing Reading",
                                     "% Overall Passing Rate"]]
```
